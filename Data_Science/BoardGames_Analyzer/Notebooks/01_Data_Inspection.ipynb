```python
# Cell 1: environment check
import sys
print("Python:", sys.version)

# Cell 2: list CSV files in data/raw
import pathlib, os
data_dir = pathlib.Path("data/raw")
csvs = sorted([p for p in data_dir.glob("**/*.csv")])
print(f"Found {len(csvs)} CSV files under {data_dir.resolve()}")
for p in csvs:
    print("-", p.relative_to(data_dir), f"({p.stat().st_size//1024} KB)")

# Cell 3: preview a few likely files (adjust filenames if different)
import pandas as pd
sample_files = csvs[:6]  # first 6 CSVs found
for p in sample_files:
    print("\n--", p.name)
    try:
        df = pd.read_csv(p, nrows=5)
        print(df.columns.tolist())
        display(df.head(3))
    except Exception as e:
        print("Could not read:", e)
```
