```python
import sys
print("Python:", sys.version)

import pathlib, os
data_dir = pathlib.Path("../Dataset/Raw")
csvs = sorted([p for p in data_dir.glob("**/*.csv")])
print(f"Found {len(csvs)} CSV files under {data_dir.resolve()}")
for p in csvs:
    print("-", p.relative_to(data_dir), f"({p.stat().st_size//1024} KB)")

import pandas as pd
sample_files = csvs[:6]
for p in sample_files:
    print("\n--", p.name)
    try:
        df = pd.read_csv(p, nrows=5)
        print(df.columns.tolist())
        display(df.head(3))
    except Exception as e:
        print("Could not read:", e)
```

```python
import pandas as pd

files = [
    "large_bgg-26m-reviews.csv",
    "large_games_detailed_info2025.csv",
    "games.csv",
    "games_detailed_info.csv",
    "mechanics.csv",
    "themes.csv",
    "subcategories.csv"
]

for f in files:
    print("\n=== Preview:", f, "===")
    path = f"../Dataset/Raw/{f}"
    try:
        df = pd.read_csv(path, nrows=5)
        print(df.columns.tolist())
        display(df.head(3))
    except Exception as e:
        print("Error reading", f, ":", e)
```
